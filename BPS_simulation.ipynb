```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

a)

```python
N = 100
n_loop = 100
x = np.zeros((N, 1))
MSD = np.zeros((n_loop, N))
t = [i for i in range(N)]

for i in range(n_loop):
    for j in range(1,N):
      dx = np.random.choice([-1, 1])
      x[j] = x[j-1] + dx
    
    for k in range(N):
      sum = 0 
      for l in range(k, N):
        sum = sum + np.square(x[l] - x[l-k])
      MSD[i, k]  = sum/(N - k)
MSD_avg =  np.sum(MSD, axis = 0)/n_loop 
plt.figure()
plt.plot(t,MSD_avg)
plt.xlabel("Time step")
plt.ylabel("MSD")
plt.figure()
plt.plot(np.log(t),np.log(MSD_avg) )
slope, intercept = np.polyfit(np.log(t[2:N]),np.log(MSD_avg[2:N]), 1)
plt.title("Slope- " + str(slope))
plt.legend(["log MSD vs log t"])
plt.xlabel("ln(Time step)")
plt.ylabel("ln(MSD)")
```

```python
N = 100       #No of timesteps
N2 = 100      #No of ensembles
x = np.zeros((N, 1))
MSD_1 = np.zeros((N2, N))
dx = np.zeros((N, 1))
t = [i for i in range(N)]

w = 0.9      #Probability

for i in range(N2):
    dx[1] = int(np.random.choice([-1, 1]))
    x[1] = dx[1]
    for j in range(2, N):
      c = int(dx[j-1])
      dx[j] = np.random.choice([c, -c], p=[w, 1-w])
      x[j] = x[j-1]+dx[j]

    
    for k in range(N):
      sum = 0 
      for l in range(k, N):
        sum = sum + np.square(x[l] - x[l-k])
      MSD[i, k]  = sum/(N - k)
MSD_avg =  np.sum(MSD, axis = 0)/N2
plt.figure()
plt.plot(t,MSD_avg)
plt.xlabel("Time step")
plt.ylabel("MSD")
plt.figure()
plt.plot(np.log(t),np.log(MSD_avg) )
slope, intercept = np.polyfit(np.log(t[2:N]),np.log(MSD_avg[2:N]), 1)
plt.title("Slope- " + str(slope))
plt.legend(["log MSD vs log t"])
plt.xlabel("ln(Time step)")
plt.ylabel("ln(MSD)")
```
